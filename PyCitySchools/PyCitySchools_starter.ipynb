```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_d = "Resources/schools_complete.csv"
student_d = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_d = pd.read_csv(school_d)
student_d = pd.read_csv(student_d)

# Combine the data into a single dataset
school_com = pd.merge(student_d, school_d, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_school = school_d["School ID"].count()
total_student = school_d["size"].sum()
total_bud = school_d["budget"].sum()
avg_mscore = student_d["math_score"].mean()
avg_rscore = student_d["reading_score"].mean()
math_pass = student_d[student_d["math_score"] >= 70 ].count()["math_score"] * 100 / total_student
read_pass = student_d[student_d["reading_score"] >= 70 ].count()["reading_score"] *100 / total_student
over_pass = (avg_mscore + avg_rscore) / 2
dis_summary = pd.DataFrame(
              {"Total Schools": [total_school],
               "Total Students": [total_student],
               "Total Budget": [total_bud],
               "Average Math Score": [avg_mscore],
               "Average Reading Score": [avg_rscore],
               "% Passing Math": [math_pass],
               "% Passing Reading": [read_pass],
               "% Overall Passing Rate": [over_pass]
              })
dis_summary["Total Students"] = dis_summary["Total Students"].map("{:,}".format)
dis_summary["Total Budget"] = dis_summary["Total Budget"].map("${:,.2f}".format)
dis_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
per_bud = school_d["budget"] / school_d["size"]
school_g = student_d.groupby(["school_name"])
avg_mscore_s = school_g["math_score"].mean()
avg_mscore_r = school_g["reading_score"].mean()
s_count = school_g["school_name"].count()
math_pass_g = student_d[student_d["math_score"] >= 70 ].groupby(["school_name"]).count()["math_score"] * 100 / s_count
read_pass_g = student_d[student_d["reading_score"] >= 70 ].groupby(["school_name"]).count()["reading_score"] * 100 / s_count
avg_pass_g = (math_pass_g + read_pass_g) / 2
sch_summary_l = pd.DataFrame(
             {"School Name": school_d["school_name"],
              "School Type": school_d["type"],
              "Total Students": school_d["size"],
              "Total School Budget": school_d["budget"],
              "Per Student Budget": per_bud,
              })
sch_summary_r = pd.DataFrame({
              "Total Students": s_count,
              "Average Math Score": avg_mscore_s,
              "Average Reading Score": avg_mscore_r,
              "% Passing Math": math_pass_g,
              "% Passing Reading": read_pass_g,
              "% Overall Passing Rate": avg_pass_g
})
sch_summary = pd.merge(sch_summary_l, sch_summary_r, on="Total Students",how="left")
sch_summary.set_index("School Name", inplace = True)
sch_summary["Total School Budget"] = sch_summary["Total School Budget"].map("${:,.2f}".format)
sch_summary["Per Student Budget"] = sch_summary["Per Student Budget"].map("${:,.2f}".format)
sch_summary
```
